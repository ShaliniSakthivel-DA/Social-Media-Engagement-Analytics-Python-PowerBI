Task 4 — Data Wrangling

```python
# Creating Additional DataFrame

country_df = df[['country']].drop_duplicates()

country_df['country_code'] = range(1, len(country_df)+1)

country_df.head()
```

```python
# Merge DataFrames

df = df.merge(
    country_df,
    on='country',
    how='left'
)

df.head()
```

```python
# Creating Engagement Score

df['engagement_score'] = (
    df['likes'] +
    df['comments'] +
    df['shares']
)

df[['likes','comments','shares','engagement_score']].head()
```

```python
# Log Transformation of Likes

df['log_likes'] = np.log1p(df['likes'])

df[['likes','log_likes']].head()
```

```python
# Hashtag Count

df[['hashtags','hashtag_count']].head()
```

```python
# Groupby
# Summary by Post Type

post_type_summary = df.groupby('post_type').agg(
    avg_likes=('likes','mean'),
    avg_comments=('comments','mean'),
    avg_shares=('shares','mean'),
    avg_engagement=('engagement_score','mean')
)
post_type_summary
```

```python
# Summary by Country

country_summary = df.groupby('country').agg(
    avg_impressions=('impression_count','mean'),
    avg_engagement_rate=('engagement_rate','mean')
)
country_summary
```

```python
# Summary by Sentiment

sentiment_summary = df.groupby('sentiment').agg(
    avg_likes=('likes','mean'),
    avg_comments=('comments','mean'),
    avg_engagement=('engagement_score','mean')
)
sentiment_summary
```

* Comments & Shares: Both 'comments' and 'shares' show much tighter distributions compared to 'likes', with lower medians and smaller IQRs. This suggests that the number of comments and shares is generally much lower and less varied than likes, with a few outliers for exceptionally high counts.
* Engagement Rate: The 'engagement_rate' box plot is highly compressed and near zero, with many extreme outliers extending far above the main box. This indicates that while most posts have a very low engagement rate, a small number of posts achieve extraordinarily high engagement rates.
* Overall Skewness & Outliers: All four metrics show a positive skew, with the majority of data points clustered towards the lower end and a tail extending to higher values. The presence of numerous outliers, especially in 'engagement_rate', suggests that a few highly successful posts significantly impact the overall average for these metrics.

```python
# Scatter: likes vs impressions

plt.figure(figsize=(8,5))
plt.scatter(
    df['impression_count'],
    df['likes']
)
plt.title("Likes vs Impressions")
plt.xlabel("Impressions")
plt.ylabel("Likes")
plt.show()
```

Interpretation of the 'Likes vs Impressions' scatter plot:

* No Clear Linear Relationship: There is no discernible positive or negative linear trend between impression_count and likes. The points are widely scattered across the plot.
* Wide Range of Likes for All Impressions: Posts with both low and high impression counts can receive a wide range of likes, from very few to nearly 20,000.
* Impressions Don't Guarantee Likes: A high number of impressions (reaching a large audience) does not automatically translate into a high number of likes, as posts with high impressions can still have low likes.

```python
# Line: daily engagement trend

daily_engagement = df.groupby(
    df['posted_at'].dt.date
)['engagement_rate'].mean()
plt.figure(figsize=(10,5))

plt.plot(
    daily_engagement.index,
    daily_engagement.values
)
plt.title("Daily Engagement Trend")
plt.xlabel("Date")
plt.ylabel("Engagement Rate")
plt.xticks(rotation=45)
plt.show()
```

Interpretation the 'Daily Engagement Trend' chart:

* High Volatility: The daily engagement rate shows considerable fluctuation, with sharp peaks and drops occurring frequently throughout the observed period, indicating high day-to-day variability.
* Sporadic Peaks: There are several distinct spikes where the engagement rate reaches very high values (e.g., nearing 25, then around 15, and other smaller peaks), suggesting that certain days or events drive exceptionally high engagement.
* No Clear Long-term Trend: Over the roughly two-year period (early 2022 to late 2023), there isn't a strong, consistent increasing or decreasing long-term trend; the engagement rate generally oscillates around a lower baseline with occasional surges.
* Baseline Engagement: For most days, the engagement rate hovers at relatively low levels (often below 5), indicating that while there are occasional high-engagement days, typical daily engagement is modest.

```python
# Swarm plot: engagement vs device

plt.figure(figsize=(8,5))
sample_df = df.sample(100)
sns.swarmplot(
    data=sample_df,
    x='device_type',
    y='engagement_rate',
    size=3
)
plt.title("Engagement vs Device")
plt.xlabel("Device Type")
plt.ylabel("Engagement Rate")
plt.show()
```

Interpretation: 'Engagement vs Device' swarm plot (based on a sample of the data):

* Low Base Engagement Across All Devices: For all three device types (desktop, mobile, tablet), the majority of posts exhibit very low engagement rates, with most data points clustered near the bottom of the plot, indicating that high engagement is not common.
* Outliers Present for All Devices: Each device type shows several data points extending to higher engagement rates. These represent individual posts that achieved significantly better engagement, regardless of the device they were posted from.
* Tablets Show Highest Individual Engagement: Although less dense with points, the 'tablet' category displays some of the highest individual engagement rate outliers, reaching above 7.0, suggesting that some content shared via tablets can achieve exceptional virality.
* Similar Distribution for Desktop and Mobile (Lower Range): In the lower engagement range, the distributions for 'desktop' and 'mobile' appear broadly similar, implying that for the bulk of content, the device type doesn't drastically alter the likelihood of low to moderate engagement.
